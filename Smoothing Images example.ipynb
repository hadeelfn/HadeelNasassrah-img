```python
import cv2
import numpy as np
from matplotlib import pyplot as plt 
     
img = cv2.imread('ff.png') 
     
kernel = np.ones((13,11),np.float32)/9 #keep this kernel  pixel, add all the 9 pixels below this kernel, take its average and replace the central pixel with the new average value. It continues this operation for all the pixels in the image
dst = cv2.filter2D(img,-1,kernel)
     
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Image Blurring (Image Smoothing)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('ff.png')
blur = cv2.blur(img,(9,9)) #size of mask
    
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('ff.png')
blur = cv2.GaussianBlur(img,(3,3),0)
    
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('ff.png')
median = cv2.medianBlur(img,5)
    
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('ff.png')

blur = cv2.bilateralFilter(img,9,75,75)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('blured')
plt.xticks([]), plt.yticks([])
plt.show()
```
